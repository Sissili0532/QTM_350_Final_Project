```python
import wbgapi as wb
import pandas as pd
import sqlite3
import os

# 0. Define your directory
DATA_DIR = '/Users/sissili/Desktop/QTM350/QTM_350_Final_Project/QTM350_Final_Project_Data' 
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)

# 1. Define the indicators
indicators = {
    'SP.ADO.TFRT': 'adolescent_fertility',
    'SE.PRM.ENRR': 'primary_education',
    'SE.SEC.ENRR': 'secondary_education',
    'SE.TER.ENRR': 'tertiary_education',
    'NY.GDP.PCAP.CD': 'gdp_per_capita'
}

# 2. Fetch data (2020-2024)
# columns='series' makes indicators the columns; labels=True adds country names
df_raw = wb.data.DataFrame(indicators.keys(), time=range(2020, 2025), labels=True, columns='series').reset_index()

# 3. Clean up the column names
# FIX: Changed 'df.rename' to 'df_raw.rename'
df_raw = df_raw.rename(columns={
    'economy': 'country_code',
    'CountryName': 'Country',
    'time': 'year',
    'NY.GDP.PCAP.CD': 'gdp_per_capita',
    'SE.PRM.ENRR': 'primary_education',
    'SE.SEC.ENRR': 'secondary_education',
    'SE.TER.ENRR': 'tertiary_education',
    'SP.ADO.TFRT': 'adolescent_fertility'
})

# 4. Save to SQLite Database in your Project Directory
db_path = os.path.join(DATA_DIR, 'final_project_database.db')
conn = sqlite3.connect(db_path)
df_raw.to_sql('wdi_raw', conn, index=False, if_exists='replace')
conn.close()

# 5. Save a CSV version in the Project Directory
csv_path = os.path.join(DATA_DIR, 'wdi_raw_data.csv')
df_raw.to_csv(csv_path, index=False)

print(f"Success! Raw data saved to:")
print(f"- Database: {db_path}")
print(f"- CSV: {csv_path}")

# 6. Preview the first 10 rows
df_raw.head(10)
```
